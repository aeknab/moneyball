```python
import pandas as pd
import os

# Load the merged_matchdays.csv file
merged_df = pd.read_csv('/Users/andreas/Desktop/02 - Moneyball/01 - Streamlit/Moneyball/data/merged_matchdays.csv')

# Define the directory where all matchday files are stored
matchdays_directory = '/Users/andreas/Desktop/02 - Moneyball/Just the Tipp/Matchdays/'

# Loop through all matchday files and update the merged_df
for matchday in range(1, 35):
    # Load the matchday file
    matchday_file = os.path.join(matchdays_directory, f'matchday_{matchday}.csv')
    matchday_df = pd.read_csv(matchday_file)
    
    # Filter FCU matches in the merged dataframe for the specific matchday
    fcu_matches = merged_df[(merged_df['Matchday'] == matchday) & ((merged_df['Home Team'] == 'FCU') | (merged_df['Away Team'] == 'FCU'))]
    
    # Iterate through the FCU matches and update the predictions
    for index, match in fcu_matches.iterrows():
        if match['Home Team'] == 'FCU':
            # FCU is the home team
            match_column = f"FCU - {match['Away Team']}"
        else:
            # FCU is the away team
            match_column = f"{match['Home Team']} - FCU"
        
        # Find the row in matchday_df corresponding to this match
        for _, player_row in matchday_df.iterrows():
            player_name = player_row['Name']
            try:
                predicted_score = player_row[match_column]
                
                # Ensure that the predicted_score is a string
                if isinstance(predicted_score, str):
                    if match['Home Team'] == 'FCU':
                        merged_df.at[index, f'{player_name} Home Goals Predicted'], merged_df.at[index, f'{player_name} Away Goals Predicted'] = predicted_score.split(':')
                    else:
                        merged_df.at[index, f'{player_name} Away Goals Predicted'], merged_df.at[index, f'{player_name} Home Goals Predicted'] = predicted_score.split(':')
                else:
                    print(f"Unexpected data type in matchday {matchday} for player {player_name}: {predicted_score}. Skipping.")
            except KeyError:
                print(f"Match {match_column} not found in matchday {matchday} for player {player_name}. Skipping.")

# Display the updated DataFrame section with FCU matches
updated_fcu_matches = merged_df[(merged_df['Home Team'] == 'FCU') | (merged_df['Away Team'] == 'FCU')]
print(updated_fcu_matches)
```

```python
def update_fcu_predictions(merged_df, matchday_df, matchday):
    fcu_matches = merged_df[(merged_df['Home Team'] == 'FCU') | (merged_df['Away Team'] == 'FCU')]
    
    for index, match in fcu_matches.iterrows():
        if match['Home Team'] == 'FCU':
            home_team, away_team = 'FCU', match['Away Team']
        else:
            home_team, away_team = match['Home Team'], 'FCU'
        
        match_column = f"{home_team} - {away_team}"
        
        for _, player_row in matchday_df.iterrows():
            player_name = player_row['Name']
            predicted_score = player_row.get(match_column)
            
            if pd.notna(predicted_score):
                if match['Home Team'] == 'FCU':
                    merged_df.at[index, f'{player_name} Home Goals Predicted'], merged_df.at[index, f'{player_name} Away Goals Predicted'] = predicted_score.split(':')
                else:
                    merged_df.at[index, f'{player_name} Away Goals Predicted'], merged_df.at[index, f'{player_name} Home Goals Predicted'] = predicted_score.split(':')

    return merged_df
```

```python
import pandas as pd
import os

# Load the merged_matchdays.csv file
merged_df = pd.read_csv('/Users/andreas/Desktop/02 - Moneyball/01 - Streamlit/Moneyball/data/merged_matchdays.csv')

# Inspect the DataFrame to ensure it loaded correctly
print(merged_df.head())
```

```python
# Define the directory containing the matchday files
matchday_dir = '/Users/andreas/Desktop/02 - Moneyball/Just the Tipp/Matchdays/'

# Create a list to store DataFrames for each matchday
matchday_dfs = []

# Loop through all 34 matchday files and rename Rheinergie to Hermann
for i in range(1, 35):
    file_path = os.path.join(matchday_dir, f'matchday_{i}.csv')
    
    # Load the current matchday file
    matchday_df = pd.read_csv(file_path)
    
    # Replace 'Rheinenergie' with 'Hermann' in the 'Name' column
    matchday_df['Name'] = matchday_df['Name'].replace('Rheinenergie', 'Hermann')
    
    # Append the DataFrame to the list
    matchday_dfs.append(matchday_df)

# Check the first matchday DataFrame to ensure the renaming worked
print(matchday_dfs[0].head(8))
```

```python
# First, we'll create a copy of the merged_df to avoid modifying the original DataFrame directly
updated_merged_df = merged_df.copy()

# Loop through each matchday DataFrame and update the FCU games in the merged DataFrame
for i, matchday_df in enumerate(matchday_dfs, start=1):
    # Filter the FCU matches in the merged DataFrame for the current matchday
    fcu_matches = updated_merged_df[(updated_merged_df['Matchday'] == i) & 
                                    ((updated_merged_df['Home Team'] == 'FCU') | 
                                     (updated_merged_df['Away Team'] == 'FCU'))]
    
    # Update predictions for each player in these matches
    for index, match in fcu_matches.iterrows():
        if match['Home Team'] == 'FCU':
            # FCU is the home team
            home_team, away_team = 'FCU', match['Away Team']
        else:
            # FCU is the away team
            home_team, away_team = match['Home Team'], 'FCU'
        
        # Find the corresponding column in the matchday DataFrame
        match_column = f"{home_team} - {away_team}"
        
        # Loop through each player and update the predictions in the merged DataFrame
        for _, player_row in matchday_df.iterrows():
            player_name = player_row['Name']
            try:
                predicted_score = player_row[match_column]
                if match['Home Team'] == 'FCU':
                    updated_merged_df.at[index, f'{player_name} Home Goals Predicted'], \
                    updated_merged_df.at[index, f'{player_name} Away Goals Predicted'] = predicted_score.split(':')
                else:
                    updated_merged_df.at[index, f'{player_name} Away Goals Predicted'], \
                    updated_merged_df.at[index, f'{player_name} Home Goals Predicted'] = predicted_score.split(':')
            except KeyError:
                print(f"Match {match_column} not found in matchday {i} for player {player_name}. Skipping.")

# Inspect the updated DataFrame to ensure the changes were applied
print(updated_merged_df[(updated_merged_df['Home Team'] == 'FCU') | 
                        (updated_merged_df['Away Team'] == 'FCU')].head())
```

```python
# Loop through all FCU games in the updated_merged_df
for index, match in updated_merged_df[(updated_merged_df['Home Team'] == 'FCU') | 
                                      (updated_merged_df['Away Team'] == 'FCU')].iterrows():
    
    # Determine if FCU is the home or away team
    is_home = match['Home Team'] == 'FCU'
    
    # Loop through each player and determine the result
    for player in ['Andreas', 'Gerd', 'Geri', 'Hermann', 'Johnny', 'Moddy', 'Samson']:
        home_goals_pred = match[f'{player} Home Goals Predicted']
        away_goals_pred = match[f'{player} Away Goals Predicted']
        
        if home_goals_pred > away_goals_pred:
            result = 'Home Win' if is_home else 'Away Win'
        elif home_goals_pred < away_goals_pred:
            result = 'Away Win' if is_home else 'Home Win'
        else:
            result = 'Tie'
        
        # Update the Result Predicted column
        updated_merged_df.at[index, f'{player} Result Predicted'] = result

# Inspect the DataFrame to ensure the changes were applied
print(updated_merged_df[(updated_merged_df['Home Team'] == 'FCU') | 
                        (updated_merged_df['Away Team'] == 'FCU')][[f'{player} Result Predicted' for player in ['Andreas', 'Gerd', 'Geri', 'Hermann', 'Johnny', 'Moddy', 'Samson']]].head())
```

```python
# Filter for predictions made by Johnny and Geri
johnny_predictions = updated_merged_df[['Johnny Home Goals Predicted', 'Johnny Away Goals Predicted']]
geri_predictions = updated_merged_df[['Geri Home Goals Predicted', 'Geri Away Goals Predicted']]

# Create a function to check if a prediction is 2:1 or 1:2
def is_2_1_or_1_2(home_goals, away_goals):
    return (home_goals == 2 and away_goals == 1) or (home_goals == 1 and away_goals == 2)

# Apply the function to Johnny's predictions
johnny_2_1_or_1_2 = johnny_predictions.apply(lambda row: is_2_1_or_1_2(row['Johnny Home Goals Predicted'], row['Johnny Away Goals Predicted']), axis=1)

# Apply the function to Geri's predictions
geri_2_1_or_1_2 = geri_predictions.apply(lambda row: is_2_1_or_1_2(row['Geri Home Goals Predicted'], row['Geri Away Goals Predicted']), axis=1)

# Calculate the percentage for Johnny
johnny_percentage = johnny_2_1_or_1_2.mean() * 100

# Calculate the percentage for Geri
geri_percentage = geri_2_1_or_1_2.mean() * 100

# Display the results
print(f"Percentage of games where Johnny predicted 2:1 or 1:2: {johnny_percentage:.2f}%")
print(f"Percentage of games where Geri predicted 2:1 or 1:2: {geri_percentage:.2f}%")
```
